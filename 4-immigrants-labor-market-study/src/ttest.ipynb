```python
import pandas as pd
from scipy.stats import ttest_ind
from scipy import stats
```

```python
file_path = r'C:\Users\sharafi\Dropbox\Goethe\Immigrants\Data\data4.xlsx'
data = pd.read_excel(file_path)
```

```python
# generating immigrants dummy
data['new_Q2'] = pd.to_numeric(data['new_Q2'], errors='coerce')
data['new_Q3'] = pd.to_numeric(data['new_Q3'], errors='coerce')

data['img'] = pd.NA
data.loc[(data['new_Q2'] == 2) & (data['new_Q3'] == 2), 'img'] = 1

data.loc[(data['new_Q2'] == 1) | (data['new_Q3'] == 1), 'img'] = 0

print(data['img'].value_counts())
```

```python
# Function to perform t-tests and report means, stds, and n for both groups
def ttests(data, group_col, group_1_value, group_0_value, variables):
  
    group_0 = data[data[group_col] == group_0_value]
    group_1 = data[data[group_col] == group_1_value]
    
    results = {}
    
    for var in variables:
        group_0_clean = group_0[var].dropna()
        group_1_clean = group_1[var].dropna()

        mean_0, std_0, n_0 = group_0_clean.mean(), group_0_clean.std(), group_0_clean.count()
        mean_1, std_1, n_1 = group_1_clean.mean(), group_1_clean.std(), group_1_clean.count()

        t_stat, p_value = stats.ttest_ind(group_0_clean, group_1_clean)
        
        results[var] = {
            'group_0': {'mean': mean_0, 'std': std_0, 'n': n_0},
            'group_1': {'mean': mean_1, 'std': std_1, 'n': n_1},
            't-stat': t_stat,
            'p-value': p_value
        }
    
    return results

variables_to_test = ['Q26r1', 'Q36r1', 'Q43r1', 'Q44r1','Q45_a', 'Q45_b', 'Q47', 'Q47_1', 'Q48r1', 'Q48_1r1']  # Add the variables you want to test

ttest_results = ttests(data, group_col='img', group_1_value=1, group_0_value=0, variables=variables_to_test)

for var, res in ttest_results.items():
    print(f"Variable: {var}")
    print(f"Group 0 (immigrant = 0): mean = {res['group_0']['mean']}, std = {res['group_0']['std']}, n = {res['group_0']['n']}")
    print(f"Group 1 (immigrant = 1): mean = {res['group_1']['mean']}, std = {res['group_1']['std']}, n = {res['group_1']['n']}")
    print(f"T-statistic: {res['t-stat']}, P-value: {res['p-value']}")
    print('-' * 40)
```
