```python
## Try to contain both immigrant and Native

def compare_origin(df):
    # Split data
    # df_img = df_female
    # df_nat = df_male
    df_female = df.loc[df['Gender'] == 'Female']
    df_male = df.loc[df['Gender'] == 'Male']
    # COunt obs 
    df_female = df_female.loc[df_female['Gender']!= 'Prefer not to say']
    df_male = df_male.loc[df_male['Gender']!= 'Prefer not to say']
    # Count obs
    count = np.array([
        df_male['img'].value_counts().loc[1],
        df_female['img'].value_counts().loc[1],
                       ])
    # Observations 
    nobs = np.array([
        len(df_male),
        len(df_female)])

    fraction = (count / nobs)*100
    z_stat, p_value = sm.stats.proportions_ztest(count, nobs, alternative='two-sided')

    p_value = '<0.001' if p_value < 0.001 else f'{p_value:.3f}'
    rslt = pd.DataFrame({'Male': f'{fraction[0]:.2f}', 'Female': f'{fraction[1]:.2f}', 'p-value': p_value},index=['Immigrant'])

    return rslt

rslt1 = compare_origin(df)
rslt1
```

```python
def compare_age(df):
    # Split data
    #df['new_age'] = df['Age'].replace({'35-49 years':'35 years or older','50 years or older':'35 years or older'})
    df_female = df.loc[df['Gender'] == 'Female']
    df_male = df.loc[df['Gender'] == 'Male']
    
    # Observations 
    nobs = np.array([
        len(df_male),
        len(df_female)])

    rslt = pd.DataFrame({'Male':[],'Female':[],'p-value':[]})
    for age in ['20-24 years','25-29 years','30-34 years','35 years or older']:
        # Count obs
        count = np.array([
            df_male['new_age'].value_counts().loc[age],
            df_female['new_age'].value_counts().loc[age],
                       ])
        fraction = (count / nobs)*100
        z_stat, p_value = sm.stats.proportions_ztest(count, nobs, alternative='two-sided')

        p_value = '$<$0.001' if p_value < 0.001 else f'{p_value:.3f}'
        new_row = pd.DataFrame({'Male': f'{fraction[0]:.2f}', 'Female': f'{fraction[1]:.2f}', 'p-value': p_value},index=[age])
        rslt = pd.concat([rslt,new_row])

    return rslt

rslt4 = compare_age(df)
rslt4
```

```python
def compare_study_field(df):
    # Split data
    df_female = df.loc[df['Gender'] == 'Female']
    df_male = df.loc[df['Gender'] == 'Male']
    
    # Observations 
    nobs = np.array([
        len(df_male),
        len(df_female)])

    rslt = pd.DataFrame({'Male':[],'Female':[],'p-value':[]})
    for study_field in ['Business & Economics','Engineering & CS','Natural & Life Sciences']:
        # Count obs
        new_name = study_field.replace('&','\&')
        count = np.array([
            df_male['StudyFieldCluster'].value_counts().loc[study_field],
            df_female['StudyFieldCluster'].value_counts().loc[study_field],
                       ])
        fraction = (count / nobs)*100
        z_stat, p_value = sm.stats.proportions_ztest(count, nobs, alternative='two-sided')

        p_value = '$<$0.001' if p_value < 0.001 else f'{p_value:.3f}'
        new_row = pd.DataFrame({'Male': f'{fraction[0]:.2f}', 'Female': f'{fraction[1]:.2f}', 'p-value': p_value},index=[new_name])
        rslt = pd.concat([rslt,new_row])

    return rslt

rslt5 = compare_study_field(df)
rslt5
```

```python
def compare_degree(df):
    # Split data
    df_female = df.loc[df['Gender'] == 'Female']
    df_male = df.loc[df['Gender'] == 'Male']
    
    # Observations 
    nobs = np.array([
        len(df_male),
        len(df_female)])

    rslt = pd.DataFrame({'Male':[],'Female':[],'p-value':[]})
    for degree in ['Bachelor','Master','PhD']:
        # Count obs
        count = np.array([
            df_male['DegreeCompact'].value_counts().loc[degree],
            df_female['DegreeCompact'].value_counts().loc[degree],
                       ])
        fraction = (count / nobs)*100
        z_stat, p_value = sm.stats.proportions_ztest(count, nobs, alternative='two-sided')

        p_value = '$<$0.001' if p_value < 0.001 else f'{p_value:.3f}'
        new_row = pd.DataFrame({'Male': f'{fraction[0]:.2f}', 'Female': f'{fraction[1]:.2f}', 'p-value': p_value},index=[degree])
        rslt = pd.concat([rslt,new_row])

    return rslt

rslt6 = compare_degree(df)
rslt6
```

```python
def compare_grad_year(df):
    df['new_grad'] = df['GradYear'].replace({2018:2020,2019:2020})
    # Split data
    df_female = df.loc[df['Gender'] == 'Female']
    df_male = df.loc[df['Gender'] == 'Male']
    
    # Observations 
    nobs = np.array([
        len(df_male),
        len(df_female)])

    rslt = pd.DataFrame({'Male':[],'Female':[],'p-value':[]})
    for grad_year in range(2020,2025):
        # Count obs
        count = np.array([
            df_male['new_grad'].value_counts().loc[grad_year],
            df_female['new_grad'].value_counts().loc[grad_year],
                       ])
        fraction = (count / nobs)*100
        z_stat, p_value = sm.stats.proportions_ztest(count, nobs, alternative='two-sided')

        p_value = '$<$0.001' if p_value < 0.001 else f'{p_value:.3f}'
        grad_year = f'{grad_year} or earlier' if grad_year == 2020 else f'{grad_year}'
        new_row = pd.DataFrame({'Male': f'{fraction[0]:.2f}', 'Female': f'{fraction[1]:.2f}', 'p-value': p_value},index=[grad_year])
        rslt = pd.concat([rslt,new_row])

    return rslt

rslt7 = compare_grad_year(df)
rslt7
```

```python
def generate_latex_table(rslt,rslt1,rslt4,rslt5,rslt6,rslt7,rslt8, filename="balance_table.tex"):
    """
    Generate a LaTeX table from a DataFrame with numerical values rounded
    to two decimal places and p-values to three decimal places.
    
    Parameters:
    df (pd.DataFrame): The data to format.
    filename (str): The output .tex file name.
    """
    # LaTeX Table Header
    latex_code = r"""
        \begin{tabular}{lccc}
            \toprule
             & Male &  Female & p-value \\
            \midrule
    """
    # Add rows dynamically
    for locs, row in rslt.iterrows():
        latex_code += f"      \\textbf{{{locs}}} & {row['Male']} & {row['Female']} &  \\\ \n"
    latex_code += f"  \\textbf{{Gender}} &  & &  \\\ \n"
    for locs, row in rslt1.iterrows():
        latex_code += f"\\phantom{{00}}        {locs} & {row['Male']}\% & {row['Female']}\% & {row['p-value']} \\\ \n"
    latex_code += f"  \\textbf{{Region}} &  & &  \\\ \n"
    for locs, row in rslt4.iterrows():
        latex_code += f"\\phantom{{00}}        {locs} & {row['Male']}\% & {row['Female']}\% & {row['p-value']} \\\ \n"
    latex_code += f"  \\textbf{{Degree}} &  & &  \\\ \n"
    for locs, row in rslt6.iterrows():
        latex_code += f"\\phantom{{00}}        {locs} & {row['Male']}\% & {row['Female']}\% & {row['p-value']} \\\ \n"
    latex_code += f"  \\textbf{{Field of Study}} &  & &  \\\ \n"
    for locs, row in rslt5.iterrows():
        latex_code += f"\\phantom{{00}}        {locs} & {row['Male']}\% & {row['Female']}\% & {row['p-value']} \\\ \n"
    latex_code += f"  \\textbf{{Graduation Year}} &  & &  \\\ \n"
    for locs, row in rslt7.iterrows():
        latex_code += f"\\phantom{{00}}        {locs} & {row['Male']}\% & {row['Female']}\% & {row['p-value']} \\\ \n"
    for locs, row in rslt8.iterrows():
        latex_code += f"        \\textbf{{{locs}}} & {float(row['Male']):.2f} & {float(row['Female']):.2f} & {float(row['p-value']):.3f} \\\ \n"
    latex_code += f"  \\textbf{{Personality}} &  & &  \\\ \n"
    for locs, row in rslt9.iterrows():
        latex_code += f"\\phantom{{00}}        {locs} & {float(row['Male']):.2f} & {float(row['Female']):.2f} & {row['p-value']} \\\ \n"
    # LaTeX Table Footer
    latex_code += r"""        \bottomrule
    \end{tabular}
    """
    
    # Save to .tex file
    with open(filename, "w") as f:
        f.write(latex_code)
    
    return filename

dir = new_dir + "\\img_labormarket\\analysis\\Gender analysis\\Charts and tables\\descriptives.tex"
generate_latex_table(rslt,rslt1,rslt4,rslt5,rslt6,rslt7,rslt8, filename=dir)
```
