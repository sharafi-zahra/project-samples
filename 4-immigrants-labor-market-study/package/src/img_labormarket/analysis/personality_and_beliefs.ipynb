```python
df, devs = likert_comparison(df, 'Q52r',9)

rslt = native_img_comparison(df, devs)

rename_dict = {

'Q52r1_devs':'Employers prefer employees who are similar to them.',
'Q52r2_devs':'Employers generally perceive immigrants as less productive.',
'Q52r3_devs':'Immigrants have higher financial pressure.',
'Q52r4_devs':'Immigrants have lower language skills.',
'Q52r5_devs':'Immigrants know less well how \n to prepare an application and how job interviews work',
'Q52r6_devs':'Immigrants use different platforms and ways to search',
'Q52r7_devs':'Immigrants do not (re-)negotiate their wage',
'Q52r8_devs':'Immigrants have fewer connections within the industry.',
'Q52r9_devs':'Immigrants are less flexible regarding their residency.',
}

rslt.rename(index=rename_dict, inplace=True)
rslt[['Immigrant Avg', 'Native Avg', 'p-value']]
rslt['Immigrants'] = rslt['Immigrant Avg']
rslt['Natives'] = rslt['Native Avg']
rslt[['Immigrants', 'Natives', 'p-value']]

rslt.sort_values(by='Immigrants', inplace=True, ascending=False)

fig, ax = plt.subplots(figsize=(16, 10))

questions = rslt.index  # Assuming rslt is the results DataFrame
ax.axvline(0, color='gray', linestyle='--', alpha=0.6, linewidth=2)  # Reference line at zero
# Plot lines connecting all native points
ax.plot(rslt["Natives"], range(len(questions)), 'b-', alpha=0.9,color=green, linewidth=5, label=f"Natives (N={int(rslt['Native N'].min())})", zorder=2)
ax.scatter(rslt["Natives"], range(len(questions)), color=green, s=100, linewidth=3.5, zorder=3)

# Plot lines connecting all immigrant points
ax.plot(rslt["Immigrants"], range(len(questions)), 'r-', alpha=0.9,color=red, linewidth=5, label=f"Immigrants (N={int(rslt['Immigrant N'].min())})", zorder=2)
ax.scatter(rslt["Immigrants"], range(len(questions)), color=red, s=100, linewidth=3.5, zorder=3)

# Add p-values to the right of each question
for i, p_value in enumerate(rslt["p-value"]):
    p_text = f"p = {p_value:.3f}" if p_value >= 0.001 else "p < 0.001"
    ax
    t(rslt["Natives"].max() + 0.1, i, p_text, va='center', fontsize=18)

# Formatting axes and labels
ax.set_yticks(range(len(questions)))
ax.set_yticklabels(questions, fontsize=22)
ax.set_xlim(-.85,1.2)
ax.set_xlabel("Demeaned Coefficient", fontsize=22)
ax.set_xticks(np.arange(-1, 1.2, 0.2))
ax.set_xticklabels([f"{tick:.1f}" for tick in ax.get_xticks()], fontsize=18)
ax.tick_params(axis='x', labelsize=18)
ax.legend(fontsize=18, loc='lower left')
#ax.grid(linestyle='--', alpha=0.6)
#plt.savefig(BLD / 'figures' / 'reasons_beliefs.pdf', bbox_inches='tight')
# Show the updated plot
plt.show()
```
